匯入csv資料

```python
train = pd.read_csv('train.csv',encoding = "utf-8") #加上encoding = "utf-8"避免字串無法讀取
test = pd.read_csv('test.csv',encoding = "utf-8")
submission = pd.read_csv('submission.csv',encoding = "utf-8")
```

```python
train = train[train['type']!=4] #切分離群值type4避免影響建模
```

```python
from sklearn.model_selection import train_test_split
x=train[['花瓣寬度','花瓣長度','花萼寬度','花萼長度']]
y=train['type']
#將測試數據的大小設置為完整數據集的30%
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=100)
x.sample(10)
```

資料標準化

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
```

```python
x_train_std[0:5] #訓練結果標準化
```

```python
x_test_std[0:5] #測試結果標準化
```

```python
y_test[0:5]
```

建立模型

```python
import xgboost as xgb
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore') #關閉警告
xgb=xgb.XGBClassifier(n_estimators=2000, max_depth=4).fit(x_train, y_train)
print(metrics.classification_report(y_test, xgb.predict(x_test))) #顯示xgb成績
print("Score:", xgb.score(x_test, y_test))
```

```python
print(xgb.feature_importances_)
```

```python
#繪製長條圖觀察哪個項目最具影響力
from xgboost import plot_importance
plot_importance(xgb, )
plt.show()
```

```python
pred = xgb.predict(test[['花瓣寬度','花瓣長度','花萼寬度','花萼長度']])
```
